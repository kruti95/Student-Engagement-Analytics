```python
import pandas as pd

# Replace 'Dataset_Excelerate.xlsx' with the actual file name
df = pd.read_excel('Dataset_Excelerate.xlsx', sheet_name='Sheet1')

# Preview the dataset
df.head()
```

```python
df.info()
```

```python
df.head()
```

```python
date_cols = ['Learner SignUp DateTime', 'Opportunity End Date',
             'Apply Date', 'Opportunity Start Date', 'Date of Birth']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna(subset=['Institution Name'])  # Drop rows with missing institution
df['Opportunity Start Date'].fillna(pd.Timestamp('1900-01-01'), inplace=True)  # Fill missing start date
```

```python
df['Gender'] = df['Gender'].str.title().str.strip()
df['Country'] = df['Country'].str.title().str.strip()
df['Current/Intended Major'] = df['Current/Intended Major'].str.title().str.strip()
```

```python
df.to_excel('Cleaned_Dataset_Excelerate.xlsx', index=False)
```
